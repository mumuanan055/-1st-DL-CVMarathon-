```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ

import requests 
import json

popdcardurl=requests.get("https://www.dcard.tw/_api/forums/pet/posts?popular=true")
nonpopdcardurl=requests.get("https://www.dcard.tw/_api/forums/pet/posts?popular=false") #éç†±é–€
response_pop=popdcardurl.text
response_pop

respone_nonpop=nonpopdcardurl.text

data=json.loads(response_pop)
 #æ¯ä¸€ç­†è³‡æ–™åŒ…å«çš„æ¬„ä½
data[0]

#å›å‚³å¹¾ç­†è³‡æ–™ï¼Œä½¿ç”¨lenå»è¨ˆç®—
print("å›å‚³å¤šå°‘ç­†è³‡æ–™:",len(data))
print("ä¸€ç­†è³‡æ–™çš„æ¬„ä½åç¨±",data[0].keys())

```

```python
# å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€
# title,createdAt,'commentCount''likeCount'


from io import StringIO
import pandas as pd
pop_frame=pd.read_json(StringIO(response_pop))  #pdç‰ˆæœ¬éé«˜éœ€è¦ä½¿ç”¨StringIOè§£æ±º
cols=["æ¨™é¡Œ","ç™¼æ–‡æ™‚é–“","ç•™è¨€äººæ•¸","æŒ‰è®šäººæ•¸"] #å®šç¾©æ¬„ä½

pop_df=pop_frame[["title","createdAt","commentCount","likeCount"]] 
pop_df.columns=cols #å®šç¾©pop_dfçš„æ¬„ä½æœƒæ˜¯=cols
pop_df

```
