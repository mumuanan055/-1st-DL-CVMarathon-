### 比較一下範例檔案中的「r.text」與「json.loads(r.text)」讀出來的內容有什麼差異

```python
import requests
import json

url = "https://api.github.com/events"

response= requests.get(url)
a = response.text
a

#r.text為所有資料的彙整，
```

```python
jsonsample=json.loads(a)
jsonsample[0]

#使用json.loads可以將r.text檔做整理，將內容呈現類似dict的形式（key:value）
```
